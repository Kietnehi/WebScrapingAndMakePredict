```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

# Khởi tạo Chrome WebDriver và tự động tải ChromeDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Khởi tạo danh sách lưu trữ kết quả
years = list(range(2024, 2018, -1))
years2=[2018,2017,2016,2015]
all_matches = []

for year in years2:
    # Điều hướng đến trang web standings
    driver.get(standings_url)
    time.sleep(5)  # Đợi trang tải hoàn chỉnh

    # Lấy HTML sau khi trang đã tải
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Tìm bảng standings
    standings_table = soup.select('table.stats_table')[0]
    
    # Lấy các liên kết đến các đội bóng
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    # Lấy liên kết tới mùa giải trước
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    # Duyệt qua các đội bóng
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        
        # Điều hướng đến trang của đội bóng
        driver.get(team_url)
        time.sleep(5)  # Đợi trang tải hoàn chỉnh

        # Lấy HTML và sử dụng pandas để đọc bảng "Scores & Fixtures"
        html = driver.page_source
        matches = pd.read_html(html, match="Scores & Fixtures")[0]

        # Tìm các liên kết liên quan đến shooting
        soup = BeautifulSoup(html, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        
        if links:
            # Điều hướng đến liên kết shooting
            driver.get(f"https://fbref.com{links[0]}")
            time.sleep(5)  # Đợi trang tải hoàn chỉnh

            # Lấy HTML và sử dụng pandas để đọc bảng "Shooting"
            html2 = driver.page_source
            shooting = pd.read_html(html2, match="Shooting")[0]
            shooting.columns = shooting.columns.droplevel()  # Loại bỏ cấp độ của cột

            # Kết hợp dữ liệu của hai bảng (Scores & Fixtures và Shooting)
            try:
                team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
            except ValueError:
                continue

            # Lọc theo giải đấu "Premier League"
            team_data = team_data[team_data["Comp"] == "Premier League"]

            # Thêm dữ liệu vào kết quả tổng
            team_data["Season"] = year
            team_data["Team"] = team_name
            all_matches.append(team_data)

        time.sleep(1)  # Đợi 1 giây trước khi chuyển đến đội tiếp theo

# Đóng trình duyệt sau khi hoàn tất
driver.quit()

# Sau khi duyệt qua tất cả các đội, bạn có thể kết hợp tất cả dữ liệu lại

```
