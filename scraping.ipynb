```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

# Khi to Chrome WebDriver v t ng ti ChromeDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Khi to danh sch lu tr kt qu
years = list(range(2024, 2018, -1))
years2=[2018,2017,2016,2015]
all_matches = []

for year in years2:
    # iu hng n trang web standings
    driver.get(standings_url)
    time.sleep(5)  # i trang ti hon chnh

    # Ly HTML sau khi trang  ti
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Tm bng standings
    standings_table = soup.select('table.stats_table')[0]
    
    # Ly cc lin kt n cc i bng
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    # Ly lin kt ti ma gii trc
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    # Duyt qua cc i bng
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        
        # iu hng n trang ca i bng
        driver.get(team_url)
        time.sleep(5)  # i trang ti hon chnh

        # Ly HTML v s dng pandas  c bng "Scores & Fixtures"
        html = driver.page_source
        matches = pd.read_html(html, match="Scores & Fixtures")[0]

        # Tm cc lin kt lin quan n shooting
        soup = BeautifulSoup(html, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        
        if links:
            # iu hng n lin kt shooting
            driver.get(f"https://fbref.com{links[0]}")
            time.sleep(5)  # i trang ti hon chnh

            # Ly HTML v s dng pandas  c bng "Shooting"
            html2 = driver.page_source
            shooting = pd.read_html(html2, match="Shooting")[0]
            shooting.columns = shooting.columns.droplevel()  # Loi b cp  ca ct

            # Kt hp d liu ca hai bng (Scores & Fixtures v Shooting)
            try:
                team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
            except ValueError:
                continue

            # Lc theo gii u "Premier League"
            team_data = team_data[team_data["Comp"] == "Premier League"]

            # Thm d liu vo kt qu tng
            team_data["Season"] = year
            team_data["Team"] = team_name
            all_matches.append(team_data)

        time.sleep(1)  # i 1 giy trc khi chuyn n i tip theo

# ng trnh duyt sau khi hon tt
driver.quit()

# Sau khi duyt qua tt c cc i, bn c th kt hp tt c d liu li

```
